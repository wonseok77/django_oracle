```python
import cx_Oracle as ora
import pandas as pd
```

```python
# 서버 위치정보
dsn = ora.makedsn("localhost", 1521, 'orcl')
dsn
```

```python
# 커넥션 정보
# dbdb가 뭔가요
conn = ora.connect("busan_06","dbdb", dsn)
conn
```

```python
# 자동차 생성
cursor = conn.cursor()
cursor
```

```python
sql = """ Select * From Cart """
sql
```

```python
# 조회명령
cursor.execute(sql)
```

```python
# 가져온거 다 출력
row = cursor.fetchall()
row
```

```python
colname = cursor.description
colname
```

```python
cursor.close()
```

```python
conn.close()
```

```python
# 컬럼 뽑아오기
col = []
for i in range(len(colname)):
    col.append(colname[i][0])
col
```

```python
coll =[]
for tp in colname:
    coll.append(tp[0])
coll
```

```python
pd.DataFrame(row, columns=col)
```
